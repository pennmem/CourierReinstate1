```python
import pandas as pd; pd.set_option('display.max_columns', None)
import cmlreaders as cml
import numpy as np
```

```python
df = cml.get_data_index()
df['experiment'].unique()
```

```python
df = df[df['experiment'] == 'CourierReinstate1']
df[:10]
```

```python
all_evs = []

for _, row in df.iterrows():
    subject = row['subject']
    session = row['session']
    reader = cml.CMLReader(subject=subject, experiment='CourierReinstate1', session=session)
    evs = reader.load('events')
    all_evs.append(evs)

all_evs = pd.concat(all_evs)
word_evs = all_evs[all_evs['type'] == 'WORD']
```

```python
word_count = word_evs.groupby('item').size()
rec_count = word_evs[word_evs['recalled'] == 1].groupby('item').size()
df2 = pd.DataFrame({'presented': word_count, 'recalled': rec_count})
df2['recall_rate'] = df2['recalled'] / df2['presented']
pd.set_option('display.max_rows', None)
```

```python
df2.sort_values(by='recall_rate', ascending=False)
```

```python
df2.to_csv('recall_rates_by_word.csv')
```
